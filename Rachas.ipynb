**Determinar la secuencias de 0 y 1, contar el número de rachas pbservadas $C_0$**

Así:

si $x_i \leq x_i-1$ se asigna 0

si $x_i > x_i-1$ se asigna 1




```python
def convertir_rachas(numeros):

  lista = []
  if not numeros:
    return lista
  lista.append(0)
  rachas = 0


  for i in range(1, len(numeros)):
    if numeros[i] > numeros[i-1]:
      lista.append(1)
    else:
      lista.append(0)
    if lista[i] == 1 and lista[i-1] == 0:
      rachas += 1
  return lista, rachas
```

```python
result_list = convertir_rachas(numeros)
print(result_list)
```

**Calcular valor esperado, varianza y el estadisitico $Z_0$**

Como $C_0$ sigue una distribución normal de media $\mu_{c_0}$ y varianza $σ^2_{c_0}$ y $Z_0$ sigue una distribución normal (0,1). El intervalo de confianza de aceptación es:

$-Z_{α,2} \leq Z_0 \leq Z_{α,2} $

```python
def prueba_de_independencia(rachas, alpha):
  valor_esperado = (2*len(numeros)-1)/3
  varianza = (16*len(numeros)-29)/90
  Z_0 = rachas - valor_esperado / mt.sqrt(varianza)

  z_aplha = norm.ppf(1-alpha/2)

  if Z_0 > z_aplha or Z_0 < -z_aplha:
    print("No se rechaza la hipotesis nula")
  else:
    print("Se rechaza la hipotesis nula")
```
